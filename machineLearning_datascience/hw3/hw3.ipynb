```python
import numpy as np
from scipy.cluster.vq import kmeans,kmeans2
import matplotlib.pyplot as plt
```

```python
### functions ###
def construct_mat():
    '''
        Constructs matrix A and B for question 1

        PARAMETERS
        ----------
        mat : str, 'A' or 'B' is expected

        OUTPUT
        -----
        A : 150 x 150 np.array, 3x3 block matrix
    
    '''

    p = 0.7
    q = 0.3

    # not sure if this is the most efficient way to generate A
    A_diag = np.ones((50,50))*p
    A_offdiag = np.ones((50,50))*q
        
    A_1 = np.hstack((A_diag,A_offdiag,A_offdiag))
    A_2 = np.hstack((A_offdiag,A_diag,A_offdiag))
    A_3 = np.hstack((A_offdiag,A_offdiag,A_diag))
    A = np.vstack((A_1,A_2,A_3))
        
    B = np.random.rand(150,150)

    # might be a faster way than for loops, check this out
    for i in range(0,150):
        for j in range(0,150):
            if B[i,j] > A[i,j]:
                A[i,j] = 1
            else:
                A[i,j] = 0

    # definitely didn't need to generate A as a block matrix, could compare
    # B[i,j] to 0.3 or 0.7 depending on what i and j are equal to!
    return A
```

```python
def random_permutation(A):
    '''
        Generates random permutation to apply to A in Q1
    '''

    # need to permute rows AND columns (does this work?...)
    #rng = np.random.default_rng()
    #A = rng.permutation(A,axis=0)
    #A = rng.permutation(A,axis=1)

    # need to use A random permutation
    #I = np.eye(150)
    idx = np.arange(150)
    np.random.shuffle(idx) # shuffles idx

    # might be a better way to do this, idk
    P = np.zeros((150,150))

    print(type(P))
    print(type(idx))
    for i in range(0,150):
        P[i,idx[i]] = 1

    # permute rows and columns
    A_rand = P @ A @ P.T

    # might need to return the permutation matrix as well?
    return A_rand,P
```

```python
# function to call to compute the mean squared error
def compute_centroid_dist(A,centroids):
    '''
        Function to compute kmeans mean squared distance
    
        PARAMETERS
        ----------
        A : observation matrix
        centroids : output of scipy kmeans2
        
        OUTPUT
        -----
        err : sum of mean squared distance
    '''
    
    # I need to learn how to use broadcasting lol cant keep coding like this
    err = 0
    #print(err)
    #print(len(centroids))
    for i in range(0,150):
        label = centroids[1][i]
        # check magnitude of this error
        err = err + np.sum((A[i,:] - centroids[0][label]) ** 2)
    
    return err
    
```

```python
# check whether or not the clusters we found are correct
def check_clusters(A,clusters):
    '''
        Checks whether or not the clusters found are appropriate by reversing the permutations.
    
    
        PARAMETERS
        ----------
        A : np.array (n x n). randomly permutated matrix
        clusters : kmeans clusters and labels generated from rows of A
        
        OUTPUTS
        -------
        B : A ordered by label
    '''
    
    labels = clusters[-1]
    # un-permute
    
    n_clusters = int(np.max(labels) + 1)
    B = np.zeros(A.shape)
    
    n = 0
    # rows
    for i in range(0,n_clusters):
        label_locations = np.where(labels == i)[0]
        n_points = len(label_locations) + n
        B[n:n_points,:] = A[label_locations,:]
        n = n_points
    
    # columns
    C = np.zeros(B.shape)
    n = 0
    # rows
    for i in range(0,n_clusters):
        label_locations = np.where(labels == i)[0]
        n_points = len(label_locations) + n
        C[:,n:n_points] = B[:,label_locations]
        n = n_points
    return C
```

```python
## Q1 - b ##
A = construct_mat()
A,Pb = random_permutation(A)

ks = np.arange(1,11)
sum_squares_err = np.zeros(ks.shape[0])
    
for i in range(0,ks.shape[0]):
    centroids = kmeans2(A,ks[i],minit='++')
    #print(centroids)
    sum_squares_err[i] = compute_centroid_dist(A,centroids)
    # last value is sum of squares

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ks,sum_squares_err)
ax.set_title("Mean squared error: 1-10 clusters")
ax.set_xlabel("Clusters")
ax.set_ylabel("Error")
plt.savefig("./figs/q1b.png")
plt.show()
```

## Q1 - b: is k=3 the right number of clusters?
Yes, $k=3$ is the right number of clusters. This is evident because there is an "elbow" in the plot of total mean squared error vs. the number of clusters used. As the number of clusters approaches the number of datapoints in the dataset, the clusters will eventually become the datapoints, and the error will approach zero. We determine the appropriate number of clusters by looking at changes of shape in the error plot, as when the benefits of using more clusters becomes marginal, this indicates that fewer clusters are optimal. In the plot above, this manifests in the "elbow" - where the slope of the error plot changes dramatically relative to its change in the rest of the plot - at $k=3$.
